## A First look at Neural Network

### Loading the MNIST dateset in Keras

```python
from keras.datasets import mnist
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()
```

### Neural Network Architechure

```python
import keras
from keras import layers

## define model
model = keras.Sequential(
    [
        layers.Dense(512, activation="relu"),
        layers.Dense(10, activation="softmax")
    ]
)
```

### Compilation step

```python
model.compile(
    optimizer="adam",
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
)
```

### Preparing the image data

```python
### We need to preprocess the data to a form that the model expects
train_images = train_images.reshape((60000, 28*28))
train_images =train_images.astype('float32') / 255

test_images = test_images.reshape((10000, 28*28))
test_images = test_images.astype('float32') / 255
```

### Fitting the model

```python
model.fit(train_images, train_labels, epochs=5, batch_size=128)
```

### Using the model to make prediction

```python
test_digits = test_images[0:10]
predictions = model.predict(test_digits)
predictions[0]
```

### Evaluate model on test data

```python
tests_loss, test_acc = model.evaluate(test_images, test_labels)
```
